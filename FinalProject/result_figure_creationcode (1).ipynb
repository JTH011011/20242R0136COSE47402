```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Zero-Shot results
zero_shot = {
    "Box(P)": 0.0827,
    "R": 0.0124,
    "mAP50": 0.042,
    "mAP50-95": 0.0288
}

# Data for Limited Training results
limited_training = {
    "Box(P)": 0.289,
    "R": 0.132,
    "mAP50": 0.114,
    "mAP50-95": 0.0625
}

# Create categories and corresponding data
categories = list(zero_shot.keys())
zero_shot_values = list(zero_shot.values())
limited_training_values = list(limited_training.values())

# Create a bar chart
x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, zero_shot_values, width, label='Zero-Shot', color='skyblue')
bars2 = ax.bar(x + width/2, limited_training_values, width, label='Limited Training', color='orange')

# Add labels, title, and legend
ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Performance Metrics: Zero-Shot vs. Limited Training')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Add value labels on top of the bars
for bars in [bars1, bars2]:
    ax.bar_label(bars, fmt='%.4f', padding=3)

# Show the chart
plt.tight_layout()
plt.savefig('performance_comparison.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Data for Zero-Shot results
zero_shot = {"Box(P)": 0.0827, "R": 0.0124, "mAP50": 0.042, "mAP50-95": 0.0288}
limited_training = {"Box(P)": 0.289, "R": 0.132, "mAP50": 0.114, "mAP50-95": 0.0625}
difference = {key: limited_training[key] - zero_shot[key] for key in zero_shot.keys()}

# Create a DataFrame with rounded values
data = {
    "Metric": list(zero_shot.keys()),
    "Zero-Shot": [f"{v:.4f}" for v in zero_shot.values()],
    "After Training": [f"{v:.4f}" for v in limited_training.values()],
    "Difference": [f"+{v:.4f}" for v in difference.values()]
}
df = pd.DataFrame(data)

# Plot the DataFrame as a table
fig, ax = plt.subplots(figsize=(8, 4))
ax.axis('off')
tbl = table(ax, df, loc='center', cellLoc='center', colWidths=[0.2] * len(df.columns))
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.auto_set_column_width(col=list(range(len(df.columns))))

# Save the table as an image
plt.savefig("df_table_image_rounded.png", dpi=300, bbox_inches='tight')
plt.show()


```
